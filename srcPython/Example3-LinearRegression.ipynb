```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jan 18 16:52:18 2018

@author: Esteban García
Method: Linear Regression
"""

print(__doc__)

import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

# Use only one feature
X_train = np.array([[1],[2],[4]])
Y_train = np.array([[2],[3],[9]])

X_test  = np.array([[3],[4],[5]])
Y_test  = np.array([[5],[6],[7]])

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, Y_train)

# The coefficients
print('Slope (B1): \n', regr.coef_)
print('Offset (B0): \n', regr.intercept_)
# The mean squared error
print("Mean squared train error: %.2f"
      % np.mean((regr.predict(X_train) - Y_train) ** 2))

print("Mean squared test error: %.2f"
      % np.mean((regr.predict(X_test) - Y_test) ** 2))


# Plot outputs
plt.scatter(X_test, Y_test,  color='black')
plt.scatter(X_train, Y_train,  color='blue')
X = np.append(X_train, X_test, axis=0)
Y = regr.intercept_ + regr.coef_*X
plt.plot(X, Y, color='red')

plt.show()
```
